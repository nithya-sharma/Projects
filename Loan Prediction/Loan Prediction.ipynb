```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("on",OneHotEncoder(drop = 'first'),[2,10])],remainder='passthrough')
df = ct.fit_transform(df)
df
```

```python
df.shape
```

```python
from scipy import stats
z = np.abs(stats.zscore(df))
z
```

```python
threshold = 3
np.where(z>3)
```

```python
df1 = df[(z<=3).all(axis=1)]
df1
```

```python
df1.shape
```

```python
X = df1[:,0:14]
Y = df1[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_pred1 = lr.predict(X_test)
Y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred1)
```

```python
#Naive Bayees
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
```

```python
pipe=Pipeline([("mn",MinMaxScaler()),("naive",GaussianNB())])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred2=pipe.predict(X_test)
Y_pred2
```

```python
accuracy_score(Y_test,Y_pred2)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
```

```python
Y_pred3=knn.predict(X_test)
Y_pred3
```

```python
accuracy_score(Y_test,Y_pred3)
```

```python
#Random Forest
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_grid=GridSearchCV(estimator=RandomForestClassifier(),param_grid={'n_estimators':np.arange(1,14),'max_depth':np.arange(2,10)},cv=5)
```

```python
rf_grid.fit(X_train,Y_train)
```

```python
rf_grid.best_params_
```

```python
rf= RandomForestClassifier(n_estimators= 7,max_depth=8)
```

```python
rf.fit(X_train,Y_train)
```

```python
Y_pred4=rf.predict(X_test)
Y_pred4
```

```python
accuracy_score(Y_test,Y_pred4)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
DecisionTreeClassifier()
```

```python
dt_grid= GridSearchCV(estimator=DecisionTreeClassifier(),param_grid={'criterion':['gini','entropy'],'max_depth':np.arange(2,8)},cv=5)
```

```python
dt_grid.fit(X_train,Y_train)
```

```python
dt_grid.best_params_
```

```python
dt1= DecisionTreeClassifier(max_depth=2)
```

```python
dt1.fit(X_train,Y_train)
```

```python
Y_pred5=dt1.predict(X_test)
Y_pred5
```

```python
accuracy_score(Y_test,Y_pred5)
```

```python
from io import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt1, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
